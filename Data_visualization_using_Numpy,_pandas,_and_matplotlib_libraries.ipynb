```python
#to create pandas dataframe with months as columns and sections as row index
import pandas as pd
import matplotlib.pyplot as plt
revenues = {
      'Sections':['Electronics','Produce','Garment'],
      'January':[33000,22000,11000],
      'February':[15000,15000,7000],
      'March':[12000,19000,12000],
      'April':[16000,30000,17000],
      'May':[19000,35000,19000],
}
dataframe=pd.DataFrame(revenues)
print(dataframe)
```

```python
#to compute and print the mean of the revenues for each month
Revmean = dataframe.mean(axis=0)
print(Revmean)
```

```python
#to draw bar chart for the mean of the revenues for each month
plt.ylabel('Mean of the revenues')
plt.title('Bar chart for the mean of revenues for each month')
ax = Revmean.plot.bar(y= 'dataframe.mean', rot=0)
```

```python
#to draw pie chart for the mean of the revenues for the months
plt.title('Pie chart for the mean of revenues')
plot = Revmean.plot.pie(y='dataframe.mean',figsize=(5,5),autopct='%1.1f%%')
```

```python
#3v to compute and print the mean for each sections
df3 = dataframe.mean(axis=1)
print(df3)
```

```python
#3vi to draw a bar chart for the mean of the revenues for each section
plt.title('Bar chart for the mean of the revenues for each section')
x=('Electronics','Produce','Garment')
ax2 = df3.plot.bar(y= 'dataframe.mean',rot=0)
```

```python
#3vii to draw a pie chart for the mean of the revenues for each section
labels=['Electronics','Produce','Garment']
plot2 = df3.plot.pie(y='dataframe.mean',labels=labels, figsize=(5,5),autopct='%1.1f%%')
```

```python
#to draw stacked barchart for each sections
stkbar = dataframe.plot.bar(x='Sections',stacked=True, rot=0)
```

```python
#to draw the line graph for the cumulative mean for the monthly revenues
cumulative=dataframe.mean(axis=0).cumsum()
cumulative.plot()
plt.show()
```
